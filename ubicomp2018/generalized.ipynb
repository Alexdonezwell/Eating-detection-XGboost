```python
import os
import pandas as pd
from beyourself import settings
from IPython.display import display, HTML
import datetime
```

```python
concat_list = []
subj_list = ['P103','P105','P107','P108','P110','P114','P116','P118','P120','P121']
latex = ""

average = []

for s in subj_list:
    path = '../data/wild/{}/generalized/{}.csv'.format(s,s)
    df_result = pd.read_csv(path)
    df_result = df_result[['day', 'precision', 'recall', 'fscore']]

    row = df_result.iloc[-1]
    
    average.append(row[1:])
    
    latex += s + " & "
    
    for j in range(1,len(row)):
        item = row[j]
        if isinstance(item, str):
            latex += item
        elif isinstance(item, datetime.date):
            latex += item.strftime("%m-%d")
        else:
            latex += "{:.1f}".format(100*item)
        if j != df_result.shape[1] - 1:
            latex += " & "
    latex += " \\\\"
    latex += "\n"

    
    concat_list.append(df_result)
    print(s)
    display(df_result)
```

```python
df_average = pd.DataFrame(average)
total_ave = df_average.mean().as_matrix()

latexnew = latex + "Ave & "
for i in total_ave:
    latexnew += "{:.1f}".format(100*i)
    latexnew += " & "

print(latexnew)
```
